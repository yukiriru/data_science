```python
import re
```

## match

```python
p=re.compile('[a-z]+') # 소문자로 된 단어
m = p.match('python')

print(m)
if m:
    print('Match found:',m.group()) # group() 매치된 문자열 반환
else:
    print('No match')

p=re.compile('[a-z]+')
m = p.match('I like python')
print(m)
```

## findall

* 동일 문자열 매치일 경우 몇 번 매치되었는지 파악할 때 의미가 있다.

```python
p.findall('I like python. Do you like python?')
```

* 정규식에 의해서 매치되는 자연어가 다를 경우에 더욱 의미가 있다.

```python
original_text='life is too short'
p=re.compile('[a-z]+')

p.search(original_text) # 첫번째 매치되는 결과를 반환
```

```python
match_list = p.findall(original_text)
# 검색 결과는 매칭된 문자열들을 리스트로 반환한다.
match_list
```

```python
for match_element in match_list:
    print(match_element)
```

## finditer

```python
original_text='life is too short'
p=re.compile('[a-z]+')

result = p.finditer(original_text)
# 매칭된 결과를 'Match object' 리스트로 반환한다. 매치 오브젝트에 대한 부가 정보를 얻을 수 있다.
result
```

```python
for r in result:
    print(r)
    print(f'r.group(): {r.group()}, r.start(): {r.start()}, r.end(): {r.end()}, r.span(): {r.span()}\n')
```
