```python
from bs4 import BeautifulSoup
import pandas as pd
```

# 데이터 로드

```python
with open('2_상품랭킹.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
soup = BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())    
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find('table', id='salesRanking')

# 테이블의 행 데이터를 추출하여 데이터프레임으로 변환
rows = table.find_all('tr')
rows[:3]
```

```python
data = []
for row in rows[1:]:  # 헤더 제외하고 반복
    cols = row.find_all('td')
    data.append([ele.text.strip() for ele in cols])

df = pd.DataFrame(data, columns=['순위', '상품명', '판매량'])
df
```
