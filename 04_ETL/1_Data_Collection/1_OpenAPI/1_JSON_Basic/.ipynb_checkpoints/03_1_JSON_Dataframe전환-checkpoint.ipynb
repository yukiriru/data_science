* 전체 데이터 개별 행 모두 읽기

```python
for student in student_info['student_list']:
    for key in student:
        print(student[key], end= ' ')
    print()
```

# 판다스 데이터 프레임으로 전환

```python
import  pandas as pd
df =pd.DataFrame(student_info['student_list'])
df
```

# 파싱 결과 저장

* json 저장

```python
import json
```

```python
with open('student_info.json','w', encoding='utf8') as outfile:
    result = json.dumps(student_info, indent=4, ensure_ascii=False)
    outfile.write(result)
    print("JSON 파일 생성 완료.")
```

* DF to CSV 저장

```python
df.to_csv('student_info.csv',encoding="cp949")
```
