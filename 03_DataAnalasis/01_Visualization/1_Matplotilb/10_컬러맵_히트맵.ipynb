## 컬러맵 설정하기


pyplot 모듈은 다양한 컬러뱁을 설정하기 위한 함수 제공

### 기본 사용

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc

import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
np.random.seed(0)
arr = np.random.standard_normal((8, 100))
arr.shape
```

```python
arr
```

```python
plt.subplot(2, 2, 1)
# c: 컬러맵에 사용되는 데이터, 행 또는 열하면 각각 가로, 세로 방향으로 컬러맵 적용
# cmap: 컬러맵 적용하는 스타일
# plt.scatter(arr[0], arr[1], c=arr[1], cmap='spring')
plt.scatter(arr[0], arr[1], c=arr[1])
plt.spring()
plt.title('spring')

plt.subplot(2, 2, 2)
plt.scatter(arr[2], arr[3], c=arr[3])
plt.summer()
plt.title('summer')

plt.subplot(2, 2, 3)
plt.scatter(arr[4], arr[5], c=arr[5])
plt.autumn()
plt.title('autumn')

plt.subplot(2, 2, 4)
plt.scatter(arr[6], arr[7], c=arr[7])
plt.winter()
plt.title('winter')

plt.tight_layout()
plt.show()
```

## 히트맵

```python
# flights 데이터셋 불러오기
import seaborn as sns
flights_data = sns.load_dataset("flights")
flights_data.shape
```

```python
flights_data.head()
```

```python
# 피벗 테이블 생성 (연도별 월별 여객 수)
pivot_flights_data = flights_data.pivot_table(index='month', columns='year', values='passengers')
pivot_flights_data.head()
```

```python
# 히트맵 플로팅
plt.figure(figsize=(10, 8))  # 플롯 사이즈 설정

# 히트맵 생성
# annot=True: 이 옵션은 각 셀에 주석을 달 것인지 여부를 지정 
# 여기서는 True로 설정하여 각 셀에 값을 표시
# fmt=".1f": 이것은 주석의 형식을 지정하는 옵션
#여기서는 소수점 이하 1자리까지 표시하도록 ".1f"를 사용했습니다. 즉, 주석에는 소수점 이하 1자리까지의 부동 소수점 숫자가 표시됩니다.
heatmap = sns.heatmap(pivot_flights_data, annot=True, fmt=".1f", cmap="YlGnBu")

# 축 레이블 설정
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Flights Heatmap')  # 플롯 제목 설정

plt.show()  # 플롯 보여주기
```
