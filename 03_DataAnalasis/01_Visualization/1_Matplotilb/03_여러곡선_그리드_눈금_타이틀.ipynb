## 수평선/수직선 표시하기

### 수평선 그리기 - axhline([y위치, x시작점(퍼센트), x종료점(퍼센트)], 기타 스타일 옵션)

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
# plt.axhline(4.0, 0.5, 0.9, color='lightgray', linestyle='--', linewidth=2)
# 여기서 0.1은 x축 너비중 좌측에서 10% 지점, 0.9는 x축 너비에서 좌측에서 90% 지점
plt.show()
```

* 마이너스 값 폰트 지원

```python
plt.rcParams['axes.unicode_minus'] = False  # 유니코드 마이너스 기호 사용
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)

plt.show()
```
