# 그래프 영역 채우기

* fill_between()
    * 두 지점을 기준으로 수평방향으로 면적을 채운다.

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.2) # alpha 값이 작을 수록 투명도는 높아진다.
#plt.fill_between(x[2:4], y[2:4], alpha=0.2) # alpha 값이 작을 수록 투명도는 높아진다.
# plt.fill_between(x[:2], y[:2], alpha=0.2) # alpha 값이 작을 수록 투명도는 높아진다.
# plt.fill_between(x[:2], y[:2], alpha=0.1) # alpha 값이 작을 수록 투명도는 높아진다.

plt.show()
```

* fill_betweenx()
    * 두 좌표를 기준으로 수직 방향의 영역을 채운다.

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)
# plt.fill_betweenx(y[1:3], x[1:3], alpha=0.5)
# plt.fill_betweenx(y[:2], x[:2], alpha=0.5)

plt.show()
```

* 다각형 영역 채우기 - fill()
