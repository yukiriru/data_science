## 파이차트

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
# autopct: % 포맷 문자열의 시작을 명시, .1f: 소스점 첫째자리까지 표기, %%: %기호를 표기
plt.show()
```

### 시작각도

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

# 오른쪽(3시)을 기준으로 0도 시작
# 데이터는 반시계방향(counterclock)으로 회전
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=0)
plt.show()
```

### 회전방향

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

# counterclock=False => 반시계방향의 반대, 즉 시계방향으로 데이터 회전
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=0, counterclock=False)
plt.show()
```

### 중심에서 벗어나는 정도 설정하기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# explode = [0, 0, 0, 0.10]
# explode = [0.3, 0, 0, 0]
explode = [0, 0, 0.9, 0]
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

### 그림자

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
explode = [0.1, 0.1, 0.1, 0.1]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.show()
```

### 색상설정

- 최대값 강조

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

* 범주형 데이터 유형 별 색상으로 구분

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

### 부채꼴 스타일 지정하기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
#wedgeprops={'widt0h': 0.7, 'edgecolor': 'w', 'linewidth': 5}
#wedgeprops={'width': 0.5, 'edgecolor': 'w', 'linewidth': 5} # 도우넛 효과를 내는 가운데 원에 대한 설정
wedgeprops={'width': 0.7, 'edgecolor': 'g', 'linewidth': 12}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import font_manager,rc
# 한글 폰트
font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
```

```python
book = pd.read_csv('book_data.csv',encoding='cp949')
book.head()
```

```python
book['book_cover'] = book['book_cover'].fillna('미정의')
```

```python
book['book_cover'].value_counts()
```

```python
value_list = book['book_cover'].value_counts().values
value_list
```

```python
label_list = book['book_cover'].value_counts().index.tolist()
label_list
```

```python
plt.pie(value_list, labels=label_list, autopct='%.1f%%')
# autopct: % 포맷 문자열의 시작을 명시, .1f: 소스점 첫째자리까지 표기, %%: %기호를 표기
plt.show()
```

* 파이의 실제 값을 표시하기

```python
#개별 % = 개별 갯수/ 총합
#개별 개수 = 개별 % * 총합
plt.pie(value_list, labels=label_list, autopct=lambda pct: f"{int(pct/100.*np.sum(value_list))}")
plt.show()
```

#### 아래 형식으로 만들어 보세요.
1235개

```python
#개별 % = 개별 갯수/ 총합
#개별 개수 = 개별 % * 총합
plt.pie(value_list, labels=label_list, autopct=lambda pct: f"{int(pct/100.*np.sum(value_list))}개")
plt.show()
```
