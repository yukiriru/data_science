* 레이블 추가

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset from seaborn
iris = sns.load_dataset('iris')
iris.head()
iris['species']
color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Define colors for each species
colors = iris['species'].map(color_map)
# Plot
plt.figure(figsize=(10, 6))
plt.scatter(iris['petal_length'], iris['petal_width'], c=colors)
plt.xlabel('꽃잎 길이')
plt.ylabel('꽃잎 너비')
plt.title('붓꽃 꽃잎 길이/너비 별 분포도')
label_text = "품종\n"
for species, color in color_map.items():
    label_text += f"{species}: {color}\n"
label_text = label_text[:-1]

plt.text(0.04, 0.96, label_text, fontsize=12, verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes, 
         bbox=dict(facecolor='white', alpha=0.5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load iris dataset from seaborn
iris = sns.load_dataset('iris')

color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Define colors for each species
colors = iris['species'].map(color_map)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(iris['petal_length'], iris['petal_width'], c=colors)
plt.xlabel('꽃잎 길이')
plt.ylabel('꽃잎 너비')
plt.title('붓꽃 꽃잎 길이/너비 별 분포도')

# Adding the text legend manually
plt.text(0.04, 0.96, "* 품종", fontsize=12, verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes, 
         )

plt.text(0.04, 0.92, "- setosa", fontsize=12, color='red', verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes)

plt.text(0.04, 0.88, "- versicolor", fontsize=12, color='green', verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes)

plt.text(0.04, 0.84, "- virginica", fontsize=12, color='blue', verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes)

plt.show()
```

* 범례 옵션을 활용한 데이터 재구성

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load iris dataset from seaborn
iris = sns.load_dataset('iris')

color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Define colors for each species
colors = iris['species'].map(color_map)

# Plot
plt.figure(figsize=(10, 6))
for species, color in color_map.items():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['petal_length'], subset['petal_width'], c=color, label=species, edgecolor='k')

plt.xlabel('꽃잎 길이')
plt.ylabel('꽃잎 너비')
plt.title('붓꽃 꽃잎 길이/너비 별 분포도')

# Adding the legend
plt.legend(title='품종', loc='upper left', edgecolor='black')

plt.show()
```

### 시나리오 설정

```python
np.random.seed(0)

n = 100
x = np.random.rand(n)
y = np.random.rand(n)
area = (50 * np.random.rand(n))**2

FIRST = 0
LAST = n-1
MEDIAN = int(n/2)
X_PADDING = 0.09
Y_PADDING = 0.07

plt.scatter(x, y, s=area, c='yellow')
plt.scatter(x[FIRST], y[FIRST], s=area[FIRST], c='red')
plt.text(x[FIRST]-X_PADDING, y[FIRST]+Y_PADDING, '1호 점포', fontdict={'size': 14})

plt.scatter(x[MEDIAN], y[MEDIAN], s=area[MEDIAN], c='green')
plt.text(x[MEDIAN]-X_PADDING, y[MEDIAN]+Y_PADDING, '중간 점포 ', fontdict={'size': 14})

plt.scatter(x[LAST], y[LAST], s=area[LAST], c='blue')
plt.text(x[LAST]-X_PADDING+0.06, y[LAST]+Y_PADDING+0.06, '가장 최신 점포', fontdict={'size': 14})
plt.title('더조은동 지역별, 상점별 매출 현황')
plt.xlabel('x 좌표 (동경 126 기준)')
plt.ylabel('y 좌표 (북위 37 기준)')
plt.show()
```

## 3차원 산점도

```python
n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')

plt.show()
```
