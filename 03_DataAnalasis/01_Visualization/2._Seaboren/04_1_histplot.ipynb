# 히스토그램 그리기

* 히스토그램은 수치 단일열의 분포를 파악하기 위한 파트로 데이터 프레임에서 단일열 이름 하나만 지정하면 된다.

```python
sns.histplot(data=penguins, x='flipper_length_mm')
```

### 히스토그램 가로로 보기

```python
sns.histplot(data=penguins, y='flipper_length_mm')
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=1)
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=2)
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=2, kde=True)
```

## 그룹별 분석

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=2, kde=True, hue='species')
```

## 스타일지정

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=2, kde=True, hue='species', element='step')
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=2, kde=True, hue='species',element='poly')
```
