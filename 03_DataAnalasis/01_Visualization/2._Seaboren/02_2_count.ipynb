# 이름별 결제 건수 시각화 비교: matplotlib vs seaborn

예시 데이터를 이용해 **이름별 결제 건수(빈도)** 를 시각화하는 두 가지 방법을 비교합니다.

```
이름,포인트
홍길동,30
이순신,10
홍길동,40
홍길동,10
이순신,15
이순신,25
```

## 목표
1. **matplotlib** 를 사용하여 `value_counts()` 결과를 막대 그래프로 그리기  
2. **seaborn** 을 사용하여 동일 데이터를 `countplot`과 `barplot(사전 집계)`로 그리기

> 참고: `countplot`은 **자동으로 빈도 집계**, `barplot`은 **사용자가 집계한 값**을 넣는 차이가 있습니다.


## 0) 라이브러리 임포트 & 데이터 생성

````python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "이름": ["홍길동", "이순신", "홍길동", "홍길동", "이순신", "이순신"],
    "포인트": [30, 10, 40, 10, 15, 25]
}
df = pd.DataFrame(data)
df
````

## 1) matplotlib + value_counts()

- `value_counts()`로 **이름별 건수**를 집계합니다.
- 결과를 `plt.bar`로 시각화합니다.


````python
counts = df["이름"].value_counts()

plt.figure()
plt.bar(counts.index, counts.values)
plt.title("이름별 결제 건수 (matplotlib)")
plt.xlabel("이름")
plt.ylabel("결제 건수")
plt.show()
````

## 2) seaborn: barplot (사전 집계)

- `barplot`은 집계된 값을 넣어줘야 합니다.
- 따라서 먼저 value_counts()로 집계한 뒤 DataFrame으로 변환합니다.


````python
counts_df = df["이름"].value_counts().reset_index()
counts_df.columns = ["이름", "결제건수"]

plt.figure()
sns.barplot(x="이름", y="결제건수", data=counts_df)
plt.title("이름별 결제 건수 (seaborn - barplot, 사전 집계)")
plt.xlabel("이름")
plt.ylabel("결제 건수")
plt.show()
````

## 3) seaborn: countplot (자동 집계)

- `sns.countplot(x="이름", data=df)`는 자동으로 빈도를 계산합니다.


````python
import seaborn as sns

plt.figure()
sns.countplot(x="이름", data=df)
plt.title("이름별 결제 건수 (seaborn - countplot)")
plt.xlabel("이름")
plt.ylabel("결제 건수")
plt.show()
````

## 정리

- **matplotlib + value_counts**: 집계를 직접 하고 bar로 그린다.
- **seaborn + barplot**: 사전 집계된 값 기반, 평균/합계 등 확장성 좋음.
- **seaborn + countplot**: 집계를 자동 처리, 코드 간결.
