```python
import seaborn as sns
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
```

# Seaborn이란?

* 데이터 분석 관점에서 품질 높은 데이터 시각화를 위한 고급 그래프를 쉽게 그릴 수 있게 기능을 제공하는 라이브러리

* 환자별(subject) timepoint마다 특정 sianal정도

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

- 기본 스타일 지정

```python
sns.set(style='darkgrid')
```

* line 차트  
    *lineplot(x=[x축 열이름], y=[y축 열이름], data=[데이터소스]

 * 차트 기본 해석
     * x축의 값이 누적데이터 일 경우 중앙 실선은 평균값으로 표현
     * 중앙 실전 주변 영역은 최소값과 최대값 범위를 표현

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

# 차트에 대한 해석예

* timepoint 범위는 0에서 시작하여 17.5까지 측정되었다.
* timepoint 별 signal 수치데이터를 보여주고 있으며 timepoint의 누적값을 표시하면서 최소값, 최대값을 영역으로 표시하고 있다.
* 최대 signal은 대략 1.7 최소 signal 대략 -0,07 측정되었다.
* siganl은 대략 0.2에서 시작하여 timepoint 5까지 급상승, 그 후 timepoint 12까지 급락, 그 후 완만하게 상승하였다.
* timepoint별 signal 분포는 대략 2.5 일때 가장 좁으며 5일때 가장 다양한 signal을 보여주었다.

# 차트 해석 Tip

* 특별한 특이사힝이 있는지 확인해본다.

* 그룹별 차트 생성
    * seaborn에서 제공한느 차트에 대부분은 hue 인자 설정이 가능
    * 이를 통해 범주형 데이터 열의값을 groupby한 범주 수만큼 개별 차트를 모두 표현 가능하다.

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
```
