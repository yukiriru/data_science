# 히스토그램이란?

- 수집한 수치형 데이터의 분포를 시각적으로 표시한 그래프

### 넘파이의 random 모듈

### 1. numpy.random.rand
 0~1 사이의 값의 uniform distribution 균일 분포, 균등 분포라고 불리는 값을 return 하는 함수

* 실생활에서 균등분포 데이터의 예
    * 주사위 던지기
    * 확률 추첨게임(로또)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.rand(5) #인자가 한개로 지정시 1차원 리스트 형태로 N개 생성
```

```python
np.random.rand(2,5) #인자가 2개인 경우 2차원 리스트로 생성
```

```python
DATA_LIMIT = 100_000
```

```python
rand_data = np.random.rand(DATA_LIMIT)
rand_data
```

```python
plt.hist(rand_data)
plt.show()
```

### 2. numpy.random.randn

 Normal(Gaussian) Distribution 정규 분포, 가우시안 분포라고 불리는 값을 return 하는 함수입니다. 
 평균 0, 분산이 1인 값을 반환

* 실생활에서 발생하는 정규분포 데이터예
    * 키, 몸무게, 성적, IQ

```python
DATA_LIMIT =100
randn_data = np.random.randn(DATA_LIMIT)
# randn_data
```

```python
plt.hist(randn_data)
plt.show()
```

# 히스토그램 응용 데이터 시뮬레이션

* 대한민국 20대 성인 남자 키 

```python
HEIGHT_AVG = 172.5
VARIANCE_AMPLICATION =5
DATA_LIMIT =100_000
```

```python
aduluts_height= np.random.randn(DATA_LIMIT)*VARIANCE_AMPLICATION +HEIGHT_AVG
```

```python
plt.hist(aduluts_height, bins=100)
plt.xlabel('키')
plt.ylabel('누적개수')
plt.show()
```

남성, 여성 키 데이터 시뮬레이션

```python
M_HEIGHT_AVG = 172.5
W_HEIGHT_AVG = 159.6

a_height= (np.random.randn(DATA_LIMIT)*VARIANCE_AMPLICATION) +M_HEIGHT_AVG
b_height= (np.random.randn(DATA_LIMIT)*VARIANCE_AMPLICATION) +W_HEIGHT_AVG

VARIANCE_AMPLICATION = 7

plt.hist((a_height,b_height), bins=100, label=['성인 남자키', '성인 여자키'])
plt.xlabel('키')
plt.ylabel('누적개수')
plt.title('전국 성인남녀 키 현황(2025.8 보건 복지부 통계자료 시뮬레이션)')
plt.legend()
plt.show()
```

# 데이터 보정작업에서의 전략

* 수집한 수치형 데이터를 시각화 해보니 정규분포형 데이터임을 알게 되었다.
* 전체 데이터는 100만건이고 결측치를 발견했다.
* 결측치의 갯수를 확인한다.
* 의사결정
    * 삭제(drop) : 결측치가 굉장히 많을 경우, 도메인 데이터 전문가 판단
    * 재조사: 전체 데이터 대비 결측치가 상당히 많은 경우
    * 결측치를 보정하기 위한 단계
        * 해당 수치데이터의 평균을 알아야 한다.
        * 평균에 근거한 정규분포 데이터를 생성한다. (randn 함수를 사용하거나 평균값으로 대체한다.)
