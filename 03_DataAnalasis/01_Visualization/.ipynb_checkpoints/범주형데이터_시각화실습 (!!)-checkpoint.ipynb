# 시각화

### Bar 차트

```python
range_list = df.비고.value_counts().index.tolist()
y = np.arange(len(range_list))
values = df.비고.value_counts().values
plt.barh(y, values)
plt.yticks(y, range_list)
plt.show()
```

* 상위 카테고리 분석

```python
def categorize_note(value) :
    if pd.isna(value):  # NaN 처리
        return "기타"
    elif "도시보건과" in value:
        return "도시보건과"
    elif "보건지소" in value:
        return "보건지소"
    elif "보건행정과" in value:
        return "보건행정과"
    elif "건강증진과" in value:
        return "건강증진과"
    else:
        return "기타"
```

```python
df['비고_상위카테고리'] = df['비고'].apply(categorize_note)
```

```python
df['비고_상위카테고리'].value_counts()
```

```python
range_list = df.비고_상위카테고리.value_counts().index.tolist()
x = np.arange(len(range_list))
values = df.비고_상위카테고리.value_counts().values
plt.bar(x, values)
plt.xticks(x, range_list)
plt.show()
```

* 상세 카테고리 분석

    * 보건지소

```python
bogoen_jiso = df[df["비고_상위카테고리"] == "보건지소"]
```

```python
bogoen_jiso['비고'].unique()
```

```python
range_list = bogoen_jiso.비고.value_counts().index.tolist()
x = np.arange(len(range_list))
values = bogoen_jiso.비고.value_counts().values
plt.bar(x, values)
plt.xticks(x, range_list, rotation=-75)
plt.show()
```

### 파이차트

```python
# counterclock=False => 반시계방향의 반대, 즉 시계방향으로 데이터 회전
plt.pie(values, labels=range_list, autopct='%.1f%%', startangle=0, counterclock=False)
plt.show()
```
