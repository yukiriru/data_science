```python
def classify_business(business_name):
    """업종명을 5개 대분류 중 하나로 반환하는 함수"""
    if business_name in ['일반음식점', '카페/베이커리', '분식', '슈퍼/마트']:
        return '외식/식료품'
    elif business_name in ['학원/교육', '병원/약국']:
        return '교육/의료'
    elif business_name in ['미용/뷰티/위생', '스포츠/헬스', '도서/문화/공연', '숙박/캠핑', '레저']:
        return '뷰티/여가/문화'
    elif business_name in ['부동산/인테리어', '자동차/자전거', '주유소']:
        return '생활편의서비스'
    # 앞의 조건에 맞지 않는 나머지는 모두 '쇼핑및기타'로 분류
    else:
        return '쇼핑및기타'

# 데이터프레임에 함수 적용
df_clean['업종별분류'] = df_clean['업종명'].apply(classify_business)
df_clean
```

```python
#소재지 도로명 주소 drop
df_clean = df_clean.drop(['소재지 도로명주소','연번'], axis=1)
```

```python
mapping = {
    '집현동': '반곡동',
    '합강동': '반곡동',
    '세종동': '나성동',  
    '가람동': '한솔동',
    '산울동': '해밀동',
}

# 안전하게 복사 후 치환
df_clean = df_clean.copy()
df_clean['읍면동'] = (
    df_clean['읍면동']
      .astype(str).str.strip()
      .replace(mapping)   # 규칙 적용
)
```

* 각 가구 유형 별 세대 수 많은 5개의 행정기관

```python
df_side = pd.concat(
    [df_people_new.nlargest(5, col)[['행정기관', col]].reset_index(drop=True) 
     for col in ['1인','2인','3인','4인','5인이상']], 
    axis=1
)
df_side
```

```python
def create_subplots():
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    axes = axes.flatten()
    
    household_types = ['1인', '2인', '3인', '4인', '5인이상']
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']
    
    for i, htype in enumerate(household_types):
        # 각 가구 유형에 맞는 행정기관 컬럼 선택
        admin_col_idx = i * 2      # 0, 2, 4, 6, 8
        value_col_idx = i * 2 + 1  # 1, 3, 5, 7, 9
        
        # 실제 데이터 값들 가져오기
        admin_names = df_side.iloc[:, admin_col_idx].values  # 행정기관 이름들
        values = df_side.iloc[:, value_col_idx].values       # 세대수 값들
        
        # 막대 그래프 생성
        bars = axes[i].bar(range(5), values, color=colors[i], alpha=0.8)
        
        # 스타일링
        axes[i].set_title(f'- {htype} 가구 TOP 5 -', fontsize=20, fontweight='bold')
        axes[i].set_xticks(range(5))
        axes[i].set_xticklabels(admin_names, rotation=45, ha='right', fontsize=15)
        
        # 값 표시
        for j, value in enumerate(values):
            axes[i].text(j, value + value*0.02, f'{int(value):,}', 
                        ha='center', va='bottom', fontsize=9)
        
        axes[i].grid(axis='y', alpha=0.3)
    
    fig.delaxes(axes[5])  # 빈 subplot 제거
    fig.suptitle('가구 유형별 행정기관 세대수 분석', fontsize=30, fontweight='bold')
    plt.tight_layout()
    plt.show()

# 실행
if __name__ == "__main__":
    create_subplots()
```

      => 비즈니스 전략1
    * tip: 평균 $2 부근 집중, 고액 일부 존재
      => 비즈니스 전략2
    * size: 2인이 가장 많음 (소규모 그룹 위주)
    * 요일/시간대 분포: 주말(Sat, Sun) & Dinner 데이터가 다수
* 초기 인사이트
    * 소비 패턴
        * 주말 저녁에 결제가 집중됨
        * 대부분 2인 중심의 식사
        * 팁과 총액은 일부 고액으로 인해 편향된 분포를 보임
    * 이상치(아웃라이어)
        * 실제 의미 있는 고객군으로 판단
* 가설 검증
    * 가설1] 고액 결제는 주로 저녁 시간에 발생할 것이다.
        * 고가의 팁과 총금액은 저녁 타임에서 발생된다.
        * 총금액과 팁은 양의 상관관계를 보여주고 있다.
    * 가설2] 인원수가 높을 수록 고액 결제를 할 것이다.
        * 전체적으로 인원수에 따라 고액 결재를 많이 하나따라 고액 결재를 많이 하나 예외도 존재한다. (최고가, 최고팁은 3인석에서 발견됨)
    * 가설3] 흡연자일 수록 고액 결제를 할 것이다.
        * 흡연자와 총결제금액, tip과의 상관관계가 뚜렷히 드러나지 않는다.
    * 추가 가설1] 요일별 시간대별 매출 상관관계가 존재 할 것으로 보임
    * 추가 가설2] 성별에 따른 매출 상관관계가 존재 할 것으로 보임
