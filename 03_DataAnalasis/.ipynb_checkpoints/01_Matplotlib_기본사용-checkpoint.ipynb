* 레이블 위치 설정

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
# x축 레벨
# x축 레이블 위치 지정 (‘left’, ‘center’, ‘right’)
plt.xlabel('X-Label', labelpad=15, fontdict=font1, loc='right')
plt.xlabel('X-Label', labelpad=15, fontdict=font1, loc='center')
# y축 레벨
# y축 레이블 위치 지정 (‘bottom’, ‘center’, ‘top’)
plt.ylabel('Y-Label', labelpad=15, fontdict=font2, loc='top')
plt.ylabel('Y-Label', labelpad=15, fontdict=font2, loc='center')
plt.show()
```

* 범례 표시하기
    * 범례: 차트에 표시되는 그래픽 요소를 구분하기 위한 레이블

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }
# Step1] Label에 범례명을 입력한다.
# plt.plot([1, 2, 3, 4], [1, 4, 9, 16] , label='Price ($)')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16] , label='Price (Won)')
# x축 레벨
# x축 레이블 위치 지정 (‘left’, ‘center’, ‘right’)
plt.xlabel('X-Label', labelpad=15, fontdict=font1, loc='right')
# y축 레벨
# y축 레이블 위치 지정 (‘bottom’, ‘center’, ‘top’)
plt.ylabel('Y-Label', labelpad=15, fontdict=font2, loc='top')
# Step2] legend함수를 호출한다.
plt.legend()
plt.show()
```

* 범례 위치 지정

```python
font1 = {'family': 'serif',
         'color': 'b',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }
# Step1] Label에 범례명을 입력한다.
plt.plot([1, 2, 3, 4], [1, 4, 9, 16] , label='Price ($)')
#x축 레벨
#x축 레이블 위치 지정 (‘left’, ‘center’, ‘right’)
plt.xlabel('X-Label', labelpad=15, fontdict=font1, loc='right')
#y축 레벨
# y축 레이블 위치 지정 (‘bottom’, ‘center’, ‘top’)
plt.ylabel('Y-Label', labelpad=15, fontdict=font2, loc='top')
# Step2] legend함수를 호출한다.
# loc=(범례의 x비율, 범례의 y비율)
# 범례를 좌측 하단에 표시
# plt.legend( loc=(0.0,0.0))
#전체 x의 50%, 전체 y의 50% 위치에 범례 표시
plt.legend( loc=(0.5,0.5))
# plt.legend( loc=(1,1))
# lower/upper와 left/right의 조합으로 표현
# plt.legend(loc='lower right')
# loc옵션에 'best' 값을 지정하면 X,Y 좌표에 표시되는 차트요소를 고려해서 최적의 위치에 표시한다.
plt.legend(loc='best') 

plt.show()
```

* 범례의 열 개수 지정

# 선 종류

### 기본유형

* linestyle 속성으로 지정하기

```python
plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 4, 4], '-', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':',label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

* 선 끝 모양 지정하기

```python
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', linewidth=10,
      solid_capstyle='butt', color='C0', label='solid+butt')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='solid', linewidth=10,
      solid_capstyle='round', color='C0', label='solid+round')

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashed', linewidth=10,
      dash_capstyle='round', color='C1', label='dashed+round')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=2)
plt.show()
```
