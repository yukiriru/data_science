# 기초 통계 항목 확인
* describe() 를 활용

## 수치형(숫자) 데이터
    

* count: 전체 개수
    * mean: 평균
    * std: 표준편차(standard deviation)
    * min: 최소값
    * 25%: 1사분위수, 1st quntile
    * 50%: 2사분위수, 중앙값(median), 2st quntile
    * 75%: 3사분위수, 3rd quntile
    * max: 최대값

```python
sample_df
```

```python
sample_df.describe()
```

## 문자와 숫자가 혼합된 데이터프레임인 경우

```python
sample_df2= pd.DataFrame(
    [
        ['a',1.2,1.3,1.4],
        ['c',9.2,9.3,2.4],
        ['d',3.2,3.3,3.4],
        ['c',4.7,2.3,5.4]
    ],
    index =['a','b','c','d'], columns=['A','B','C','D']
)
sample_df2
```

```python
sample_df2.describe() # 데이터프레임에 describe() 우선적으로 수치형 열에 대한 기초통계정보를 제공한다.
```

## 문자를 포함한 열의 기초 통계 정보

* count: 전체 관측값
* unique: 문자열의 유일한 값
* top: 가장 빈도가 높은 데이터
* freq: 최빈 데이터의 개수

```python
sample_df2
```

```python
sample_df2.describe(include=['object']) # 문자타입의 열에 대한 기초 통계 내역 조회
```

## 4 분위수

quantile 은 수치 데이터를 크기 순서로 정렬했을 때 0% ~ 100% (0 ~ 1) 위치에 해당하는 숫자를 리턴해 준다.  
즉, Q1 은 25% , Q2 는 50%, Q3 는 75%, Q4 는 100%를 의미

## 기초통계값과 데이터의 분포 관계

* Mean: 평균값
* Meadian: 중위값, 2사분위수
* Mode: 최빈값(가장 높은 빈도로 발견되는 데이터)
