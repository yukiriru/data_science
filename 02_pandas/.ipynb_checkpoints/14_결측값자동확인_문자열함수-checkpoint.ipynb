```python
movies['title_Eng'].hasnans
```

* isnull() : 개별 항목에 대한 모든 결측값(NaN) 유무에 대해서 True/False를 반환

* 열에서 결측치 여부 확인  
행단위로 결측값 확인

```python
movies['actor'].isnull().head(10)
```

* 데이터프레임에서 결측치 여부 확인

```python
movies.isnull().head(3)
```

* 결측치 열만 필터링

```python
missing_columns = movies.isnull().any() # 열단위로 결측값이 1개라도 있으면 True 아니면 False
missing_columns
```

```python
column_missing_values = missing_columns [ missing_columns == True ]
column_missing_values
```

# 특정열의 값이 null인 전체 행 필터링

* step1] 특정열의 값이 null이 True인 블리언마스크용 자료구조를 만든다.

```python
movies['actor'].isnull().head(10)
```

* step2] 해당 조건으로 블리언마스크 필터링 적용

```python
movies[ movies['actor'].isnull() ].head(10)
```

```python
movies[ movies['title_Eng'].isnull() ].head(10)
```

# NaN(결측값)을 포함한 모든 행에 대한 필터링

* step1] 행단위 결측값 유무 확인을 위한 블리언 마스크 결과 테스트

```python
movies.isnull().any(axis=1) # axis=1 조건으로 결측값 유무를 행단위로 측정. 참고) 기본값은 열단위 측정
```

* step2] 행단위 결측값을 블리언 마스크 기법으로 필터링한다.

```python
movies[ movies.isnull().any(axis=1) ].head(10)
```

# 결측값을 제외한 모든 행 필터링  
결측값 처리전에 결측값이 없는 데이터를 추출하여 선행하여 분석을 하고 싶을 때  
자기자신은 바꾸지 않는다.

```python
movies.shape
```

* dropna : 결측값이 있는 데이터 삭제  
 axis=0: 행단위 삭제

```python
movies.dropna(axis=0).head(10)
```

```python
movies.head(3)
```

### 결측값이 없는 모든 행

```python
no_na_all = movies.dropna(axis=0)
no_na_all.head(10)
```

```python
movies.shape, no_na_all.shape
```

### 특정 컬럼만 결측치를 제외하도록 필터링  
데이터프레임.dropna(axis=0) 의 의미는 매우 엄격하게 결측값을 불허용하는 원칙  
결측값 불허용의 원칙을 특정열에만 적용

```python
movies.info()
```

* 마스크 기법

```python
target_cols = ['director', 'actor'] # 결측치 불허용 열 정의
movies.isnull().any(axis=1) # 행단위 모든 결측값 유무
```

```python
movies[target_cols].isnull().any(axis=1) # 팬시 검색으로 'director', 'actor' 두 가지 열에 대해서만 결측값 유무를 판단
```

```python
mask_has_nan = movies[target_cols].isnull().any(axis=1)
```

```python
movies.loc[ mask_has_nan , : ].head(60)
```

```python
movies.loc[ ~mask_has_nan, : ].head(60)
```

* subset 적용  
  subset에서 지정한 열에서 존재하는 결측치만 제거한다.

```python
movies.dropna(subset=['director','actor']).head(60)
```
