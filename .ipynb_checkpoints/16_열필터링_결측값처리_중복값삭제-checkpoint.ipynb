# 결측값 보정(Data Imputation)

* 결측값 예

```python
import numpy as np
```

```python
s = pd.Series( [0, 1, np.nan ,3] )
s
```

#### 결측치 이전의 값으로 보정  
이전 데이터가 한 동안 유지된 패턴을 보일 경우에 유효

```python
s = pd.Series(
    [0, 
     1,
     1,
     1,
     1,
     1,
     1,
     1,
     1,
     np.nan ,
     3]
)
s
```

```python
s.fillna(method='ffill')
```

```python
s.ffill()
```

#### 결측치 이후의 값으로 보정

```python
s = pd.Series( 
    [0, 
     1, 
     np.nan ,
     3,
     3,
     3,
     3,
     3,
     3,
     3
    ] 
)
s
```

```python
s.bfill()
```

## 결측값에 대한 보간 처리

### 보간(interpolate): 결측값이 발생할 때 전 후 값을 고려하여 특정 간격의 값을 넣는 방식
* 주로 선형데이터의 결측치를 처리할 때 유용하다.
    * 선형데이터: 일정한 패턴을 두고 데이터가 증가 또는 감소하는 데이터
     (예: 물가, 나이..)

```python
data = {
    'x': [2, 4, np.nan, 8, 10],
    'y': [np.nan, 10, 20, np.nan, 50]
}
df = pd.DataFrame(data)
df
```

```python
df.interpolate()
```

## 단일 값 보정

```python
# 학생 성적 데이터를 포함하는 리스트 (일부러 결측값을 넣음)
data = [
    [80, 90, 85],
    [70, np.nan, 92],  # 학생2의 영어 성적 결측
    [95, 88, np.nan]   # 학생3의 수학 성적 결측
]

# 인덱스와 컬럼 이름 지정
index = ['학생1', '학생2', '학생3']
columns = ['국어', '영어', '수학']

# 리스트를 데이터프레임으로 변환
df = pd.DataFrame(data, index=index, columns=columns)
df
```

* 학생이 시험을 응시하지 않은 경우

```python
df.fillna(0)
```

```python
df # 자기 자신을 바꾸지는 않는다.
```

```python
df['수학'].fillna(0) # 특정열의 결측값만 보정하고자 할
```

```python
df.fillna(0, inplace=True) # 자기 자신도 바꾸려면 inplace옵션 활용
df 
```

```python
df = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수", "최은지"],
    "수학": [90, np.nan, 85, 70],
    "영어": [88, 76, np.nan, 95]
})
```

* 얕은 복사(shallow copy)

```python
df_shallow_copy = df
```

```python
df_shallow_copy.fillna(0,inplace=True)
df_shallow_copy
```

```python
df
```

-깊은 복사 (Deep Copy)  
원본을 유지하고 복사본에서 작업을 할 경우: 전처리 등의 작업을 할 때 주로 사용

```python
df2 = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수", "최은지"],
    "수학": [90, np.nan, 85, 70],
    "영어": [88, 76, np.nan, 95]
})
df2
```

```python
df2_deep_copy = df2.copy()
```

```python
df2_deep_copy.fillna(0,inplace=True)
df2_deep_copy
```

```python
df2
```

* 슬라이싱 기법을 활용한 복사본

```python
df3 = pd.DataFrame({
    "이름": ["김철수", "이영희", "박민수", "최은지"],
    "수학": [90, np.nan, 85, 70],
    "영어": [88, 76, np.nan, 95]
})
df3
```

```python
df3_deep_copy_slice = df3[:]
df3_deep_copy_slice
```

```python
df3_deep_copy_slice.fillna(0,inplace=True)
df3_deep_copy_slice
```
